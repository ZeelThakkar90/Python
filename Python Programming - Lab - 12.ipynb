### 04) WAP to demonstrate the use of Pie chart.

```python
import matplotlib.pyplot as plt
labels = ['A', 'B', 'C', 'D', 'E']
sizes = [20, 30, 25, 15, 10]
colors = ['blue', 'green', 'red', 'purple', 'orange']
plt.pie(sizes, labels=labels, colors=colors, autopct=
'%1.1f%%'
,
startangle=140)
plt.title('Pie Chart Example')
plt.show()
```

### 05) WAP to demonstrate the use of Bar chart.

```python
import matplotlib.pyplot as plt
categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['pink', 'lightgreen', 'cyan', 'purple', 'orange']
plt.bar(categories, values, color=colors)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

### 06) WAP to demonstrate the use of Scatter Plot.

### 08) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]
colors = ['pink', 'lightgreen', 'cyan', 'purple', 'orange']

plt.bar(categories, values, color=colors)

for i, v in enumerate(values):
    plt.text(i, v + 1, str(v), ha='center', fontsize=12)
    
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart with Values Displayed')
plt.show()
```

### 09) WAP create a Scatter Plot with several colors in Matplotlib?

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(50) * 10
y = np.random.rand(50) * 10

colors = np.random.rand(50)

plt.scatter(x, y, c=colors, cmap='viridis', edgecolors='pink')
plt.colorbar(label='Color Intensity')
plt.xlabel('X values')
plt.ylabel('Y values')
plt.title('Scatter Plot with Several Colors')
plt.show()
```

### 10) WAP to create a Box Plot.

```python
import matplotlib.pyplot as plt
import numpy as np

data = [np.random.randn(100) * 10 + i for i in range(1, 6)]

plt.boxplot(data, patch_artist=True, boxprops=dict(facecolor='pink'))

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot Example')
plt.show()
```
